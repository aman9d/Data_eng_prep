```python
def combine_args(*numbs):
    if not all(isinstance(x, (int, float)) for x in numbs):
        raise TypeError("All arguments must be numbers")
    return sum(numbs)

# Example use:
print(combine_args(1, 2.5, 3))  # ✅ Output: 6.5
```

1. Map: Squares of a Sorted Array,  https://leetcode.com/problems/squares-of-a-sorted-array/

```python
class Solution:
    def sortedSquares(self, nums):
        a=sorted(list(map(lambda x:x*x, nums)))
        return a

nums = [-4,-1,0,3,10]
Solution.sortedSquares(nums,nums)
```

```python
class Solution:
    def findRelativeRanks(score):
        sort_score=sorted(score,reverse=True)
        rank=["Gold Medal","Silver Medal","Bronze Medal"]+[str(i) for i in range(4,len(score)+1)]
        res=[]
        for k in score:
            for i,j in zip(sort_score,rank):
                if k==i:
                    res.append(j)

        return(res)

Solution.findRelativeRanks([1,2,4,11,3])
```

```python
class Solution:
    def minSetSize( arr) :
        count={}
        for i in arr:
            if i not in count:
                count[i]=1
            else:
                count[i]+=1

        temp=sorted(count.items(),key=lambda item:item[1],reverse=True)

        min_sum=len(arr)/2
        sum_ele=0
        count_el=0

        for i,j in temp:
            sum_ele+=j
            count_el+=1
            if sum_ele>=min_sum:
                return(count_el)


```
