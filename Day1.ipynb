```python

def even_odd(num_lst):
    even=[]
    odd=[]
    sum_all=0

    for i in num_lst:
        sum_all+=i
        if i%2==0:
            even.append(i)
        else:
            odd.append(i)
    largest_even=even[0]
    for j in even:
      if largest_even<j:
        largest_even=j

    largest_odd=odd[0]
    for j in odd:
      if largest_odd<j:
        largest_odd=j

    count_of_even=len(even)
    count_of_odd=len(odd)
    avg_all=sum_all/len(num_lst)
    largest_even=largest_even #max(even)
    largest_odd=largest_odd #max(odd)


    return count_of_even,count_of_odd,sum_all,avg_all,largest_even,largest_odd


a=[2,3,8,5,1,4]
count_of_even,count_of_odd,sum_all,avg_all,largest_even,largest_odd=even_odd(a)
print('Count of even numbers:',count_of_even)
print('Count of Odd numbers:', count_of_odd)
print('Sum of all numbers:', sum_all)
print('Avg of all numbsers:', avg_all)
print('Largest even number:', largest_even)
print('Largest odd number:', largest_odd)



```

leet code 2 sum : https://leetcode.com/problems/two-sum/description/

```python
class Solution:
  def twoSum(nums, target):
      res=[]
      for i in range(len(nums)-1):
          for j in range(i+1,len(nums)):
              if nums[i]+nums[j]==target:
                res.append(i)
                res.append(j)
                return res

nums = [3,2,4]
target = 6

Solution.twoSum(nums,target)
```

```python
def count_vowels(s):
  vowels_lst=['a','e','i','o','u']
  cnt=0
  for i in s:
    if i in vowels_lst:
      cnt+=1
  return cnt

a='abcde'
count_vowels(a)
```
